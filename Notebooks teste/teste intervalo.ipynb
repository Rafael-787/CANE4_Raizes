```python
from sympy import *
from sympy import Poly
import matplotlib.pyplot as plt
import numpy as np
```

```python
intervalo = [-100,100]

x = symbols('x')

exp1 = x**3 - 9*x +3
exp2 = cos((pi*(x+1))/8) + 0.148*x - 0.9062
exp = (x-200)**2 -25

f = lambdify(x,exp,'numpy')
nsolve(exp,x,[0,9])
```

```python
x_values = np.linspace(intervalo[0], intervalo[1], 10000)
y_values = f(x_values)
```

```python
# Faz a multiplicação dos elementos
produto = y_values[1:] * y_values[:-1]

# Cria uma lista com tuplas dos intervalos das raízes
intervalos = [(lambda i: (i,i+1))(i) for i in np.where(produto < 0)[0]] 

print(intervalos)

```

```python
plt.plot(x_values,y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.ylim(intervalo[0],intervalo[1]) #condição nas configurações
for i in intervalos:
    plt.axvline(x=x_values[i[0]], color='blue', linestyle='--')
    plt.axvline(x=x_values[i[1]], color='blue', linestyle='--')
#plt.legend()
plt.grid(True)
plt.show()
```
