```python
from sympy import *
from sympy import Poly
import matplotlib.pyplot as plt
import numpy as np
```

```python
intervalo = [-100,100]

x = symbols('x')

exp1 = x**3 - 9*x +3
exp = cos((pi*(x+1))/8) + 0.148*x - 0.9062

f = lambdify(x,exp,'numpy')
nsolve(exp,x,[0,9])
```

```python
x_values = np.linspace(intervalo[0], intervalo[1], 10000)
y_values = f(x_values)
```

```python
#y_values = np.array([-1,1,0,5,2,-3])
produto = y_values[1:] * y_values[:-1]
#print(produto)
u = np.where(produto < 0)

intervalos = [(lambda i: (i,i+1))(i) for i in np.where(produto < 0)[0]]

print(intervalos)

```
