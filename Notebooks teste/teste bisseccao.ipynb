```python
from sympy import *
from sympy import Poly
import matplotlib.pyplot as plt
import numpy as np
```

```python
intervalo = [-5,10]

x = symbols('x')

exp1 = x**3 - 9*x +3
exp = cos((pi*(x+1))/8) + 0.148*x - 0.9062
```

```python
f = lambdify(x,exp,'numpy')
nsolve(exp,x,[0,9])
```

```python
a,b = intervalo
i_old = b
e = 1
u = 0
c = 0

while e >= 0.001:
    global c
    global u
    i = (a+b)/2
    c = f(i)
    #print(f"f({i} = {c})")
    if c * f(b) <0 : 
        a = i
    else:
        b = i

    e = abs((i)-i_old)/abs(i)
    i_old = i
    #print(f"a:{a}  b:{b}  e:{e}")
    u += 1

print(f"Raiz: {i}  iterações:{u}")
```

```python
x_values = np.linspace(intervalo[0], intervalo[1], 10000)
y_values = f(x_values)

plt.plot(x_values,y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.ylim(intervalo[0],intervalo[1]) #condição nas configurações
#plt.axvline(x=i, color='blue', linestyle='--', label=f'Raiz {round(i,3)}')
#plt.legend()
plt.grid(True)
plt.show()
```
